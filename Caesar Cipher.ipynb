```python
alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 
            'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z'
            'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 
            'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

print('Welcome to Caesar Cipher üòçü•∞ where you can encrypt and decrypt messages by shifting letters in the alphabet')
print('''           
 ,adPPYba, ,adPPYYba,  ,adPPYba, ,adPPYba, ,adPPYYba, 8b,dPPYba,  
a8"     "" ""     `Y8 a8P_____88 I8[    "" ""     `Y8 88P'   "Y8  
8b         ,adPPPPP88 8PP"""""""  `"Y8ba,  ,adPPPPP88 88          
"8a,   ,aa 88,    ,88 "8b,   ,aa aa    ]8I 88,    ,88 88          
 `"Ybbd8"' `"8bbdP"Y8  `"Ybbd8"' `"YbbdP"' `"8bbdP"Y8 88   
            88             88                                 
           ""             88                                 
                          88                                 
 ,adPPYba, 88 8b,dPPYba,  88,dPPYba,   ,adPPYba, 8b,dPPYba,  
a8"     "" 88 88P'    "8a 88P'    "8a a8P_____88 88P'   "Y8  
8b         88 88       d8 88       88 8PP""""""" 88          
"8a,   ,aa 88 88b,   ,a8" 88       88 "8b,   ,aa 88          
 `"Ybbd8"' 88 88`YbbdP"'  88       88  `"Ybbd8"' 88          
              88                                             
              88           

            ''')


#create a function called caesar
def caesar(start_message, shift_number, cipher_direction):
    end_message = ""
    
    if cipher_direction == 'decode':
        shift_number *= -1
            
    for character in start_message:
        #keeping symbols/numbers/space during decoding and encoding
        if character in alphabet:
            position = alphabet.index(character)
            new_position = position + shift_number
            end_message += alphabet[new_position]
            
        else:
            end_message += character
             
    print(f'The {cipher_direction}d message is {end_message}.')
        
#Asking user whether they want to continur or stop
go_on = True
while go_on:
    
    direction = input("Type 'encode' to encrypt, type 'decode' to decrypt:\n")
    text = input("Type your message:\n").lower()
    shift = int(input("Type the shift number:\n"))

#using modulus to allow decoding and encoding of text when shift is greater than 26
    shift = shift % 26

    caesar(start_message = text, shift_number = shift, cipher_direction = direction)

    option = input("Type Yes to continue or No to stop: ")
    if option == 'no':
        go_on = False
        print('Thanks for using my Encoder/Decoder.üòöüòò')
```
